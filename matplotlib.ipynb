```python
import matplotlib.pyplot as plt
%matplotlib  inline
```

```python
print(dir(plt))
```

single data set plotting - simple plot
-----------

```python
x = [34, 23, 56, 78, 96]
print(type(x))
print(len(x))
fig = plt.figure()
plt.plot(x)
plt.title('single data set plot')
plt.xlabel('length of defined data set x')
plt.ylabel('x dataset values are taken')
plt.show()
```

2 numerical data sets plotting - simple plot 
--


```python
a = [25, 35, 45, 55, 65, 75]
b = [1, 2, 3, 4, 5, 6]
fig = plt.figure()
plt.plot(a, b)
plt.grid()
plt.title('a values on x-axis and b values on y-axis')
plt.xlabel('a')
plt.ylabel('b')
plt.show()
print("_______________**********_____________________")
fig = plt.figure()
plt.plot(b, a)
plt.grid()
plt.title('b values on x-axis and a values on y-axis')
plt.xlabel('b')
plt.ylabel('a')
plt.show()
```

one numerical and one categorical data set plotting - simple plot
--------

```python
p = ['India', 'Pakistan', 'Srilanka', 'china', 'japan']
q = [10, 20 , 45, 89, 101]
fig = plt.figure()
plt.plot(p, q)
plt.grid()
plt.title('p values on x-axis and q values on y-axis')
plt.xlabel('p')
plt.ylabel('q')
plt.show()
```

categorical and numerical dataset plotting - simple plot
----

```python
p = ['India', 'India', 'Srilanka', 'china', 'japan']
q = ['abc', 'bac', 'cde','def','efg']
fig = plt.figure()
plt.plot(p, q)
plt.grid()
plt.title('p values on x-axis and q values on y-axis')
plt.xlabel('p')
plt.ylabel('q')
plt.show()
```

Bar plot 
----------

```python
x1 = [10, 20, 30, 40, 50]
x2 = [11, 22, 33, 44, 55]
fig = plt.figure(figsize=(10,5))
plt.bar(x1, x2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Bar plot for 2 datasets')
plt.grid()
plt.savefig('bar.pdf')
plt.show()
```

pie chart 
-------

```python
fig = plt.figure()
countries = ['India', 'pak', 'china', 'srilanka']
population = [30, 5, 25, 5]
plt.pie(population, labels = countries)
plt.show()
```

```python
plt.pie(population, labels = countries, explode = (0.1,0.1,0.1,0.5))
fig = plt.figure()
plt.pie(population, labels = countries, explode = (0.1,0.1,0.1,0.1))
```

scatter plot
-----------

```python
y1 = [1, 2.5, 3.6, 2, 3, 8]
y2 = [10, 30 , 40 , 50 , 60, 70]
fig = plt.figure(figsize=(10,5))
plt.scatter(y1, y2, label='y2 values')
plt.xlabel('y1')
plt.ylabel('y2')
plt.legend(loc='upper right')
plt.grid()
plt.title('Scatter plot for 2 datasets y1 and y2 ')
plt.show()
```

Area plot
--------


```python
fig = plt.figure()
a = [1, 2.5, 3.6, 2, 3, 8]
b = [10, 30 , 40 , 50 , 60, 70]
plt.stackplot(a, b, alpha=12.8)
plt.xlabel('a')
plt.ylabel('b')
plt.title('Area plot')
plt.grid()
```

```python
import numpy as np 
x = range(50)
fig = plt.figure(figsize=(10,6))
plt.plot(np.sin(x)) # sine wave 
plt.plot(np.cos(x)) # cosine wave 
plt.show()
```
