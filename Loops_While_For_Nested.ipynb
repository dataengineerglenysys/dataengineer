# Range

start Optional. An integer number specifying at which position to start. Default is 0

stop Required. An integer number specifying at which position to stop (not included).

step Optional. An integer number specifying the incrementation. Default is 1

```python
a = range(1,10,2)
b = range(1,10) # Default step is 1
c = range(5)
```

# While loop

While loop iterates through a block of statements repetitively until the given condition returns False. This means, while loop first checks the given condition and if the condition isTrue, then it executes the block of statements inside the while loop. Whenever the condition returnsFalse, it breaks out of the loop and executes the next immediate line of code.

while loop must have a increment or decrement operation, else the loop will run infinitely,

```python
x = 1 # counter set to 1

while x < 10:
    print(x)
    x += 3 # x = x + 3
```

# Infinite while loop

```python
x = 1

while x < 10:
    print(x)
```

# for loop

In Python, for loop uses the in keyword.

```python
for a in range(3):
    print(a)
```

# Nested loop

```python
for x in range(5):
    print(x)
    for y in range(100,102):
        print(y)
```
